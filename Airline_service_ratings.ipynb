## Import Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

```python
df = pd.read_csv("/content/Airline Quality Ratings.csv")
```

## Basic Exploratory Data Analysis

```python
# Visualization of Gender distribution


plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Gender', data=df, palette={'Male': 'blue', 'Female': 'red'}, legend=False)
plt.title('Gender Distribution')
plt.show()
```

```python
# Visualization of Age Distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization of Customer type distribution

custom_palette = ["blue", "red"]

plt.figure(figsize=(8, 6))
sns.countplot(x='Customer Type', data=df, palette=custom_palette)
plt.title('Customer Type Distribution')
plt.show()
```

```python
# Visualization of Type of travel distribution

plt.figure(figsize=(8, 6))
sns.countplot(x='Type of Travel', data=df)
plt.title('Type of Travel Distribution')
plt.show()
```

```python
# Visualization of Class distribution

plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
# Visualization of Satisfaction distribution

plt.figure(figsize=(8, 6))
sns.countplot(x='Satisfaction', data=df)
plt.title('Satisfaction Distribution')
plt.show()
```

```python
# Visualization of Departure delay distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['Departure Delay'], bins=30, kde=True)
plt.title('Departure Delay Distribution')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization of Arrival delay distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['Arrival Delay'].dropna(), bins=30, kde=True)
plt.title('Arrival Delay Distribution')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization of Flight distance distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['Flight Distance'], bins=30, kde=True)
plt.title('Flight Distance Distribution')
plt.xlabel('Flight Distance')
plt.ylabel('Frequency')
plt.show()
```

1.   **Gender Distribution:** The gender distribution seems fairly balanced, indicating that both males and females are likely represented in the dataset.

2.  **Age Distribution:** The age distribution appears to be somewhat skewed, with more data points concentrated within certain age ranges. Further analysis could reveal specific age groups that are more prevalent among travelers.
